# Question 1

## Generate a pandas data frame contains 10 entries with the following columns: Name [A - Z], Age [7 - 90], Gender [M / F], and City [A - Z] and fill it with randomly generated Numbers or Alphabetical Characters as shown with brackets.

```python
def generate(a = list(string.ascii_letters),n = 10,length = 10):
    a = np.array(a)
    randomChoices = np.random.choice(a,[n,length])
    result = ["".join(randomChoices[i]) for i in range(len(randomChoices))]
    return result
```

```python
data = {
    "Names" : generate(n=100),
    "Age" : np.random.randint(7,90,size=(100)),
    "Gender" : generate(a = list("MF"),n=100,length=1),
    "City" : generate(n=100)
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

## Create a new column called Category that assigns each person to a category based on their age: ”Child” for ages under 18, ”Adult” for ages 18-65, and ”Senior” for ages over 65. Then, count the number of people in each category and display the result.

```python
def classify(age):
    if age < 18:
        return "Child"
    elif age < 65:
        return "Adult"
    else:
        return "Senior"
category = []
for row in df["Age"]: category.append(classify(row))
df["Category"] = category
print(df)
df.groupby("Category").describe()
```

## Generate a pandas data frame that contains 10 rows of information about different movies, including the title [A - Z], genre[”Comedy” / ”Drama” / ”Action”], year [1990 - 2010], and rating [1 - 10].

```python
d = {
    "title" : generate(n = 100),
    "genre" : random.choices( ["Comedy","Drama","Action"],k=100),
    "year" : np.random.randint(1990,2011,size = 100),
    "rating": np.random.randint(1,10,size=100)
}
```

```python
df2 = pd.DataFrame(d)
df2
```

```python
df2[(df2["year"] >= 2000) & (df2["rating"] >= 7)].sort_values(by='rating', ascending=False).head(10)
```
